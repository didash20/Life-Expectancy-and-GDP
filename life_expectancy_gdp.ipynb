# Life Expectancy and GDP

In this project, we analyze data on GDP and life expectancy from the World Health Organization and the World Bank to try and identify the relationship between the GDP and life expectancy of six countries.

During this project, we analyze, prepare, and plot data in order to answer questions in a meaningful way.

After performing the analysis, we'll be creating a blog post to share our findings on the World Health Organization website.

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the data

```python
df = pd.read_csv('all_data.csv')
len(df)
```

```python
df.head()
```

```python
df = df.rename(columns={'Life expectancy at birth (years)':'Life_Exp'})
```

## Check Countries

```python
countries = df['Country'].unique()
countries
```

## Evaluate Country Life Expectacy by Year

```python
fig = plt.figure(figsize=(6,4))
fig.patch.set_facecolor('#F5FFF5')
ax = fig.subplots()
sns.scatterplot(data=df, x='Year',y='Life_Exp', hue='Country')
ax.title.set_text('Countries Life Expectancy by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Life expectancy at birth (years)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
plt.show()
```

## Evalueate Country GDP by Year

```python
fig = plt.figure(figsize=(6,4))
fig.patch.set_facecolor('#F5FFF5')
ax = fig.subplots()
sns.scatterplot(data=df, x='Year',y='GDP', hue='Country')
ax.title.set_text('Countries GDP by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Life expectancy at birth (years)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
plt.show()
```

## Evaluate Correlation between GDP and Life Expectancy

```python
fig = plt.figure(figsize=(6,4))
fig.patch.set_facecolor('#F5FFF5')
ax = fig.subplots()
sns.scatterplot(data=df, x='Life_Exp',y='GDP', hue='Country')
ax.title.set_text('GDP vs Life Expectancy by country')
ax.set_ylabel('GDP in Trillions of U.S. dollars')
ax.set_xlabel('Life expectancy at birth (years)')
ax.ticklabel_format(style='sci', axis='y', scilimits=(12,12))
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
plt.show()
```

```python
fig = plt.figure(figsize=(12,7))
fig.patch.set_facecolor('#F5FFF5')
colors = sns.color_palette()
for i,country in enumerate(countries):
    ax = plt.subplot(2,3,i+1)
    df_country = df[ df.Country == country ]
    sns.scatterplot(data=df_country, x='Life_Exp',y='GDP', color=colors[i])
    ax.title.set_text(country)
    ax.ticklabel_format(style='sci', axis='y', scilimits=(12,12))
    
    if( i < 3 ):
        ax.set_xlabel('')
    else:
        ax.set_xlabel('Life expectancy at birth (years)')
    
    if( i % 3 == 0):
        ax.set_ylabel('GDP in Trillions of U.S. dollars')
    else:
        ax.set_ylabel('')
        
fig.suptitle('GDP vs Life Expectancy by country',fontsize=16)
fig.tight_layout()
fig.legend(countries, loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
plt.show()
```
